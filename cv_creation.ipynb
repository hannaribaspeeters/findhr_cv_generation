```python
import pandas as pd
import numpy as np
import random

# Load data
education_institutions = pd.read_csv("./tables/tourism/education_institutions.csv")
education_degrees = pd.read_csv("./tables/tourism/education_degrees.csv")
professions = pd.read_csv("./tables/tourism/professions.csv")
eu_languages = ["spanish", "english", "italian", "french", "german"]

# Utility functions
def get_random_items(data, n):
    """Return a random sample of items from a list."""
    return random.sample(data, min(n, len(data)))

def generate_educational_background():
    """Generate the educational background section of the CV."""
    ed_items = np.random.randint(1, 4)  # Number of education entries
    educational_background = []
    
    level = random.randint(0,1)
    for i in range(ed_items):
        degrees = education_degrees.loc[education_degrees["level"]==level, "degree"].dropna().values
        degree = np.random.choice(degrees)
    
        institutions = education_institutions.loc[
                education_institutions["Type"] == ("School" if level == 0 else "University"),
                "Institution"
            ].values
        
        institution = np.random.choice(institutions)
        
        educational_background.append({
                "degree": degree,
                "start_date": "",
                "end_date": "",
                "institution": institution
            })
    
        level += 1
    
        return educational_background


def generate_professional_background(age_group):
    """Generate the professional background section of the CV."""
    role_counts = {
        1: {"entry": np.random.randint(1, 4), "mid": 0, "senior": 0},
        2: {"entry": np.random.randint(2, 4), "mid": np.random.randint(1, 4), "senior": 0},
        3: {"entry": np.random.randint(2, 4), "mid": np.random.randint(3, 4), "senior": np.random.randint(1, 4)}
    }

    professional_background = []
    h_skills, s_skills, n_languages = [], [], 0

    def add_roles(role_list, count):
        nonlocal h_skills, s_skills, n_languages
        used_roles = []
        for _ in range(count):
            role = np.random.choice(list(set(role_list) - set(used_roles))).item()
            used_roles.append(role)

            companies = professions.loc[professions["Role"] == role, "Companies"].values[0].split(",")
            professional_background.append({
                "role": role,
                "start_date": "",
                "end_date": "",
                "institution": np.random.choice(companies).item()
            })

            h_skills.extend(professions.loc[professions["Role"] == role, "Hard Skills"].values[0].split(","))
            s_skills.extend(professions.loc[professions["Role"] == role, "Soft Skills"].values[0].split(","))
            n_languages = max(n_languages, professions.loc[professions["Role"] == role, "Languages"].values[0])

    # Add roles based on age group
    counts = role_counts[age_group]
    add_roles(professions[professions["Level"] == "Entry"]["Role"].values, counts["entry"])
    add_roles(professions[professions["Level"] == "Mid"]["Role"].values, counts["mid"])
    add_roles(professions[professions["Level"] == "Senior"]["Role"].values, counts["senior"])

    return professional_background, list(set(h_skills)), list(set(s_skills)), n_languages

def generate_cv(age_group):
    """Generate the complete CV."""
    educational_background = generate_educational_background()
    professional_background, h_skills, s_skills, n_languages = generate_professional_background(age_group)

    # Select a random subset of skills and languages
    hard_skills = get_random_items(h_skills, np.random.randint(np.min([len(h_skills)-1,5]) , len(h_skills)))
    soft_skills = get_random_items(s_skills, np.random.randint(np.min([len(s_skills)-1,5]), len(s_skills)))
    languages = get_random_items(eu_languages, n_languages)

    return {
        "educational_background": educational_background,
        "professional_background": professional_background,
        "soft_skills": soft_skills,
        "hard_skills": hard_skills,
        "languages": languages
    }

# Generate the CV
age_group = 1  # Example: Entry-level age group
cv = generate_cv(age_group)
print(cv)
```
